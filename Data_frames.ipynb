### Filtering a DataFrame

```python
# Now we will apply functions and sove question using ipl data
ipl.head(2)
```

```python
# Find all final winners
mask=ipl['MatchNumber']=='Final'
new_df=ipl[mask]
new_df[['Season','WinningTeam']]
```

```python
#How many super over finishes have occured
ipl[ipl['SuperOver']=='Y'].shape
```

```python
#How many matches has csk won in kolkata
ipl[(ipl['City']=='Kolkata')& (ipl['WinningTeam']=='Chennai Super Kings')] 
```

```python
# How many matches has CSK lost in Mumbai
ipl[(ipl['City']=='Mumbai')& ((ipl['Team1']=='Chennai Super kings') | (ipl['Team2']=='Chennai Super Kings'))&(ipl['WinningTeam']!='Chennai Super Kings')]
```

```python
# Toss winner is match winner in percentage
(ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
#Movies with rating higher then 8 amd voting>1000
movies.head()
movies[(movies['imdb_rating']>8)&(movies['imdb_votes']>1000)]
```

```python
# action movies with rating higher than 7.5
movies[(movies['genres'].str.contains('Action'))&(movies['imdb_rating']>7.5)]
```
