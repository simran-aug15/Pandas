### Data Frame : It is a tabular data ,2D data consistes of rows and column.Every row is series (if taken single) and every column is series(if taken single)

```python
import numpy as np
import pandas as pd
```

```python
#creating data frame using list
student_data=[[100,80,10],[90,70,7],[120,100,4],[80,50,2]]
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
#using dict
student_dict={'iq':[100,90,120,80],
              'marks':[80,70,100,50],
              'package':[10,7,14,2]}
students=pd.DataFrame(student_dict)
students
```

```python
#using read_csv
movies=pd.read_csv('movies.csv')
movies
```

```python
#using read_csv
ipl=pd.read_csv('ipl-matches.csv')
ipl
```

### DataFrame Attribute and Methods

```python
#shape : To check the shape of the data set
movies.shape
```

```python
#dtypes : To check the type of data set
movies.dtypes
```

```python
#index : To get the index value of data set
movies.index
```

```python
#columns : To get the column value of data set
movies.columns
```

```python
#values : To get the values of the data set
students.values
```

```python
#head : To get the first top values of data set 
movies.head()
```

```python
#tail : To get the lower most values of data set
movies.tail()
```

```python
#sample : To get the random value of data set 
movies.sample()
```

```python
#describe : It provides the mathematical summary of data
movies.describe()
```

```python
#info : gives high level info about the data
movies.info()
```

```python
#isnull : check whether your data have null values or not
movies.isnull()
```

```python
#This will give you the sum of total null values in one column
movies.isnull().sum()
```

```python
#duplicated : To check the repeated data frame(False=not duplicated,true=yes,duplicated)
movies.duplicated()
```

```python
#rename : Use to rename the column name of data set
students.rename(columns={'marks':'percent','package':'lpa'})
```

### Math methods

```python
#sum : apply sum on each column
movies.sum()
students.sum()
```

```python
#sum : apply sum on axis=1 then it will be applied on row
students.sum(axis=1)
```

```python
#mean : Give u average of values
students.mean(axis=1)
```

```python
#var : give u variance of values
students.var()
```

```python
#std : gives u the standard deviation of data set
students.std()
```
