### Important Methods of DataFrame

```python
import numpy as np
import pandas as pd
```

```python
#Import the data set
ipl=pd.read_csv("ipl-matches.csv")
movies=pd.read_csv('movies.csv')
batsman=pd.read_csv('batsman_runs_ipl.csv')
diabetes=pd.read_csv('diabetes.csv')

```

```python
#value_counts : This count the number of values (Series,DataFrame)
#Example for series
a=pd.Series([1,2,1,1,2,3])
a.value_counts()
```

```python
#value_counts example for data frames
marks=pd.DataFrame([[100,80,10],[90,70,7],[120,100,14],[80,70,14],[80,70,14]],columns=['iq','marks','package'])
marks.value_counts()
```

```python
#Find which player has won most potm in finals and qualifirs 
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
#How many matches a team played
(ipl['Team1'].value_counts()+ipl['Team2'].value_counts()).sort_values(ascending=True)
```

```python
#sort_values : It sort the values in ascending order
#It workd on series aswell as DataFrame
movies.head(4)
movies.sort_values(['title_x'])
```

```python
#multiple column
movies.sort_values(['year_of_release','title_x'])
```
