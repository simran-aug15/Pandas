```python
#value_counts : This count the number of values (Series,DataFrame)
#Example for series
a=pd.Series([1,2,1,1,2,3])
a.value_counts()
```

```python
#value_counts example for data frames
marks=pd.DataFrame([[100,80,10],[90,70,7],[120,100,14],[80,70,14],[80,70,14]],columns=['iq','marks','package'])
marks.value_counts()
```
