```python
import pandas as pd
import numpy as np
```

```python
#importing data
movies=pd.read_csv('imdb-top-1000.csv')
delivers=pd.read_csv('deliveries.csv')
```

```python
#making groups on the basis of genre
genres=movies.groupby('Genre')
genres
```

### Apply builtin aggregation function on groupby object

```python
#sum()
genres.sum()
```

```python
# Find the top 3 genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
# Find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Find director which is most popular
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#Find no. of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

### Groupby Attributes And Methods

```python
# len : Find total number of groups
len(movies.groupby('Genre'))
```

```python
# size : Find items in each group
movies.groupby('Genre').size()
```

```python
# First()/last(): nth item
genres.first()    # first movies of data set
```

```python
genres.last()     # it gives the last movies
```

```python
genres.nth(6)     #Find the 6th movie in dataset
```

```python
#get_group : we can see all movies of particular group
movies[movies['Genre']=='Fantasy']
```
