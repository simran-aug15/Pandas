##Pandas : It is a fast,powerful,flexible and easy to use open source data analysis and manipulation tool,built on top ofthe Python

##Pandas series: It is like a column in table .It is a 1D array holding data of any type

```python
# importing pandas
import pandas as pd
import numpy as np
```

##Series from lists

```python
#String
country=["India","Pakistan","USA","Nepal","Srilanka"]
pd.Series(country)
```

```python
#Integer
runs=[12,13,15,17,18,19,19]
pd.Series(runs)
```

```python
#Custom index
marks=[65,78,29,98]
subjects=["Maths","English","Science","Hindi"]
pd.Series(marks,index=subjects)
```

```python
#Setting a name
pd.Series(marks,index=subjects,name="simran ke marks")
```

##Series from dict

```python
marks={"Maths":67,"English":78,"Hindi":90,"Science":89}
marks_series=pd.Series(marks,name="simran ke marks")
marks_series
```

##Series Attributes

```python
#size=Number of items in series
marks_series.size
```

```python
#dtype=type of the elemnets in series
marks_series.dtype
```

```python
#name=The we added in the name section of the code
marks_series.name
```

```python
#is_unique=It will gives true value for unique numbers in series but if it sees the repeated number than gives false
marks_series.is_unique
```

```python
#index=Gives the index values
marks_series.index
```

```python
#value=Gives numpy array
marks_series.values
```

##Series using read_csv

```python
#importing movie data
#reading the file
bollywood=pd.read_csv('bollywood.csv')
bollywood
```

```python
#making series of movie column
bollywood=bollywood['movie']
bollywood
```

```python
#making the lead column
bollywood=bollywood['lead']
bollywood
```

```python
#making lead column->index
bollywood=pd.read_csv('bollywood.csv',index_col='lead')
bollywood
```

```python
#making lead column->index
bollywood=pd.read_csv('bollywood.csv',index_col='movie')
bollywood
```
